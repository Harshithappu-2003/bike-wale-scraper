## Import libraries

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup as bs
```

```python
url="https://www.bikewale.com/best-scooters-in-india/"
response=requests.get(url)
```

```python
response.status_code
```

```python
soup=bs(response.content,"html.parser")
```

```python
print(soup.prettify())
```

```python
container=soup.find('div',{'class':'o-fzoHFO o-fzoHCA o-fzoHBq'})
print(container.prettify())
```

```python
scooters=container.find_all('li',{'class':'o-fznJzb o-fHmpzP'})
```

```python
for scooter in scooters:
  print(scooter.prettify())
```

```python
img_url=[]
names=[]
prices=[]
ratings=[]
```

```python
for scooter in scooters:
  names.append(scooter.img.get('alt'))
print(names)
```

```python
for scooter in scooters:
  img_url.append(scooter.img.get('src'))
print(img_url)
```

```python
prices=[]
for scooter in scooters:
  price=scooter.find('span',{'class':'o-cJrNdO o-byFsZJ o-bkmzIL o-bVSleT'}).text
  # print(price)
  prices.append(price)
print(prices)
```

```python
for scooter in scooters:
    rating_element = scooter.find('span', {'class': 'o-cwUopP o-fzpimR o-KxopV o-sTQWx o-dThPjR'})
    
    if rating_element:  # Ensure the element is found
        ratings = rating_element.text
        if ratings:
            rating = ratings.split()[0]
            print(rating)
```
